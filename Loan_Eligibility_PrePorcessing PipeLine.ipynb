```python
def high_credit_score(data):
    
    New_Credit_Score=[]
    for i in data['Credit Score']:
        
        if np.isnan(i):
            New_Credit_Score.append(i)
        else :
            if float(str(i)[:3])>900:
                 New_Credit_Score.append(900.0)
            else :
                 New_Credit_Score.append(float(str(i)[:3]))
    data['Credit Score'] = New_Credit_Score               
    return data
```

```python
###Home Ownership

class HomeOwnSpell(BaseEstimator,TransformerMixin):
    def __init__(self):
        self.X=X
    def fit(self,X,y=None):
        return self
    def transform(self,X,y=None):
        X.replace({'Home Ownership' :{'HaveMortgage':'Home Mortgage'}},inplace=True)
        return X
```

```python
###Purpose
class PurposeSpell(BaseEstimator,TransformerMixin):
    def __init__(self):
        self.X=X
    def fit(self,X,y=None):
        return self
    def transform(self,X,y=None):
        X.replace({'Purpose':{'other':'Other','Take a Trip':'vacation'}},inplace=True)
        return X
```
